```python
import matplotlib.pyplot as plt
import csv
import pandas as pd

x = []
y = []
```

```python
data = pd.read_csv("audio_files_data.csv", header = 0)

x = data['Duration of audio file']
y = data['Number of audio files with that duration']

#plt.barh(x, y)
plt.hist(x, bins=250, weights=y, histtype='bar', color='r')
plt.xlabel('Duration of audio file (ms)')
plt.ylabel('Number of audio files')
plt.title('Audio file duration vs number of files')
```
